# Contexte et application

## Desciption de l'hypertension

L'hypertension artérielle est une maladie cardiovasculaire. Selon l'OMS, elle est responsable de 17 millions de décès par an dans le monde. Soit environ un tiers de la mortalité totale [Source: ncbi.nlm.nih.gov].

Si l'hypertension n'est pas prise en charge, cela provoque des problèmes à long terme comme par exemple l'infractus du myocarde, l'accident vasculaire cérébral ou l'insuffisance cardiaque.

Pour réduire les risque, il faut avoir un poids normal, avoir une activité physique, limiter la consommation d'alcool et manger sain. [Source: Fondation pour la recherche Médicale]

## Alimentation pour les hypertendus et choix des variables.

### Le sucre

Une chose à laquelle il faut faire attention, c'est une alimentation trop sucrée. Il faut donc limiter cette consommation à 25g/jour. Les sucres se présentent sous plusieurs forme : glucose, galactose, fructose, saccharose, lactose, maltose, amidon. Dans notre fichier, nous choisissons ceux qui sont dans notre fichier grâce à la liste des variables fournis.
1.  glucose_100g
2.  fructose_100g
3.  lactose_100g
4.  maltose_100g
5.  starch_100g (amidon)
6.  carbohydrates_100g ( glucides)
7.  sugars_100g

[Source : ameli.fr]  

Plus un produit contiendra du sucre moins il sera recommandé pour la consommation de l'hypertendu. 


### Potassium, calcium, magnésium

Il faut favoriser la consommation du potassium, du calcium et du magnésium. [Source: Doctissimo]
Plus un individus va en consommer, à part le potassium, mieux ce sera pour sa santé.
On va donc sélectionner les variables suivantes pour notre étude:

1. potassium_100g
2. calcium_100g
3. magnesium_100g

Le potassium et le sodium aide à l'équilibre acido-basique du corps et des fluides. 
Le potassium pour les hypertendus à des effets bénéfiques entre 1,2 et 3,5 grammes par jour au dela, il peut y avoir des effets negatifs. 

### Le sel

Les hypertendus peuvent consommer du sel. La quantité:

1.  Basse est inférieur à 0.3g pour 100g d'aliment. C'est la consommation à priviléger
2.  Moyenne : de 0.3g à 1.5g pour 100g d'aliment. A consommer modèrement
3.  Haute : 1.5g ou plus pour 100g d'aliment. A éviter

Mais dans notre fichier il est noté sous forme de sodium. Pour le sodium, les quantités varient

1.  Basse est inférieur à 0.1g pour 100g d'aliment. C'est la consommation à priviléger
2.  Moyenne : de 0.1g à 0.6g pour 100g d'aliment. A consommer modèrement
3.  Haute : 0.6g ou plus pour 100g d'aliment. A éviter

[Source: ameli.fr]

On choisit donc la variable 
1.  sodium_100g


### Oméga 3, oméga 6, omega 9

Les molécules produits à partir des oméga 6 ont un effet important. Ils peuvent élever la pression arterielle.
Au contraire les omega3 ont peu d'effet. Consommer plus d'omega3 empechent les omega6 de produire leurs molécules.
[Source : estenso.org]

Les Omega3 aide à fabriquer le bon cholestérol et à dégager le mauvais. Cet effet semble décupler lors de la consommation d'acide gras polyinsaturé au détriment des acides gras saturés.[Source naturaforce.com]

Les Omega9 ont des effet bénéfique sur l'hypertension

Les mauvaises graisses sont à l'origine de l'accumulation de cholesterol dans les artères favorisant aussi l'hypertension arterielle. [Source naturaforce.com]

Le risque d'hypertension est élevé chez ceux qui mangent beaucoup d'aliments contenant des gras trans.

#### Coefficients d'asymétrie ( skweness)

1.  Un coefficient nul indique une distribution symétrique.
2.  Un coefficient négatif indique une distribution décalée à droite de la médiane, et donc une queue de distribution étalée vers la gauche.
3.  Un coefficient positif indique une distribution décalée à gauche de la médiane, et donc une queue de distribution étalée vers la droite.
[Source : Wikipédia]

````python
# On calcule le coefficient d'asymétrie de df_new_2
skewness = []
for col in df_new_2.columns:
    skewness.append(sp.stats.skew(df_new_2[col], axis=0, bias=True)) 
#skewness
data_skewness = pd.Series(skewness)
name = pd.Series(df_new_2.columns)
dt2 = {"name": name,   
      "Skewness": data_skewness
        }
dt3 = pd.concat(dt2,axis = 1) 
dt3
````

### Le Z-score

Le Z-Score est également appelé score standard. Cette valeur/score permet de comprendre à quelle distance se trouve le point de données de la moyenne. Et après avoir défini une valeur seuil, on peut utiliser les valeurs de score z des points de données pour définir les valeurs aberrantes. Si le score z d’un point de données est supérieur à 3, cela indique que le point de données est assez différent des autres points de données. Un tel point de données peut être une valeur aberrante. [Source : https://fr.acervolima.com/]

````python
z = np.abs(stats.zscore(new_data['vitamin-d_100g']))
print(z)
````

Pour définir la valeur de base des valeurs aberrantes est définie au-dessus et au-dessous de la plage normale des ensembles de données, à savoir les limites supérieure et inférieure, définissez les limites supérieure et inférieure (la valeur 1,5 * IQR est prise en compte) :

supérieur = Q3 +1,5*IQR

inférieur = Q1 – 1,5*IQR

Dans la formule ci-dessus, selon les statistiques, la mise à l’échelle de 0,5 de l’IQR (new_IQR = IQR + 0,5 * IQR) est prise en compte pour prendre en compte toutes les données comprises entre 2,7 écarts types dans la distribution gaussienne.

````python
def display_circles(pcs, n_comp, pca, axis_ranks, labels=None, label_rotation=0, lims=None):
    for d1, d2 in axis_ranks: # On affiche les 3 premiers plans factoriels, donc les 6 premières composantes
        if d2 < n_comp:

            # initialisation de la figure
            fig, ax = plt.subplots(figsize=(6,7))

            # détermination des limites du graphique
            if lims is not None :
                xmin, xmax, ymin, ymax = lims
            elif pcs.shape[1] < 30 :
                xmin, xmax, ymin, ymax = -1, 1, -1, 1
            else :
                xmin, xmax, ymin, ymax = min(pcs[d1,:]), max(pcs[d1,:]), min(pcs[d2,:]), max(pcs[d2,:])

            # affichage des flèches
            # s'il y a plus de 30 flèches, on n'affiche pas le triangle à leur extrémité
            if pcs.shape[1] < 30 :
                plt.quiver(np.zeros(pcs.shape[1]), np.zeros(pcs.shape[1]),
                   pcs[d1,:], pcs[d2,:], 
                   angles='xy', scale_units='xy', scale=1, color="grey")
                # (voir la doc : https://matplotlib.org/api/_as_gen/matplotlib.pyplot.quiver.html)
            else:
                lines = [[[0,0],[x,y]] for x,y in pcs[[d1,d2]].T]
                ax.add_collection(LineCollection(lines, axes=ax, alpha=.1, color='black'))
            
            # affichage des noms des variables  
            if labels is not None:  
                for i,(x, y) in enumerate(pcs[[d1,d2]].T):
                    if x >= xmin and x <= xmax and y >= ymin and y <= ymax :
                        plt.text(x, y, labels[i], fontsize='14', ha='center', va='center', rotation=label_rotation, color="blue", alpha=0.5)
            
            # affichage du cercle
            circle = plt.Circle((0,0), 1, facecolor='none', edgecolor='b')
            plt.gca().add_artist(circle)

            # définition des limites du graphique
            plt.xlim(xmin, xmax)
            plt.ylim(ymin, ymax)
        
            # affichage des lignes horizontales et verticales
            plt.plot([-1, 1], [0, 0], color='grey', ls='--')
            plt.plot([0, 0], [-1, 1], color='grey', ls='--')

            # nom des axes, avec le pourcentage d'inertie expliqué
            plt.xlabel('F{} ({}%)'.format(d1+1, round(100*pca.explained_variance_ratio_[d1],1)))
            plt.ylabel('F{} ({}%)'.format(d2+1, round(100*pca.explained_variance_ratio_[d2],1)))

            plt.title("Cercle des corrélations (F{} et F{})".format(d1+1, d2+1))
            plt.show(block=False)
        






````

````python
def display_factorial_planes(X_projected, n_comp, pca, axis_ranks, labels=None, alpha=1, illustrative_var=None):
    for d1,d2 in axis_ranks:
        if d2 < n_comp:
 
            # initialisation de la figure       
            fig = plt.figure(figsize=(6,7))
        
            # affichage des points
            if illustrative_var is None:
                plt.scatter(X_projected[:, d1], X_projected[:, d2], alpha=alpha)
            else:
                illustrative_var = np.array(illustrative_var)
                for value in np.unique(illustrative_var):
                    selected = np.where(illustrative_var == value)
                    plt.scatter(X_projected[selected, d1], X_projected[selected, d2], alpha=alpha, label=value)
                plt.legend()

            # affichage des labels des points
            if labels is not None:
                for i,(x,y) in enumerate(X_projected[:,[d1,d2]]):
                    plt.text(x, y, labels[i],
                              fontsize='14', ha='center',va='center') 
                
            # détermination des limites du graphique
            boundary = np.max(np.abs(X_projected[:, [d1,d2]])) * 1.1
            plt.xlim([-boundary,boundary])
            plt.ylim([-boundary,boundary])
        
            # affichage des lignes horizontales et verticales
            plt.plot([-100, 100], [0, 0], color='grey', ls='--')
            plt.plot([0, 0], [-100, 100], color='grey', ls='--')

            # nom des axes, avec le pourcentage d'inertie expliqué
            plt.xlabel('F{} ({}%)'.format(d1+1, round(100*pca.explained_variance_ratio_[d1],1)))
            plt.ylabel('F{} ({}%)'.format(d2+1, round(100*pca.explained_variance_ratio_[d2],1)))

            plt.title("Projection des individus (sur F{} et F{})".format(d1+1, d2+1))
            plt.show(block=False)
````

````python
def display_scree_plot(pca):
    scree = pca.explained_variance_ratio_*100 #  renvoie un tableau numpy unidimensionnel qui contient les valeurs du pourcentage de variance expliquée par chacune des composantes sélectionnées.
    plt.bar(np.arange(len(scree))+1, scree)
    plt.plot(np.arange(len(scree))+1, scree.cumsum(),c="red",marker='o')
    plt.xlabel("rang de l'axe d'inertie")
    plt.ylabel("pourcentage d'inertie")
    plt.title("Eboulis des valeurs propres")
    plt.show(block=False)
````

````python
def plot_dendrogram(Z, names):
    plt.figure(figsize=(10,25))
    plt.title('Hierarchical Clustering Dendrogram')
    plt.xlabel('distance')
    dendrogram(
        Z,
        labels = names,
        orientation = "left",
    )
    plt.show()
````
